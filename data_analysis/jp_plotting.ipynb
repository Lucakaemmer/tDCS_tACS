```python
import pandas as pd
import matplotlib.pyplot as plt
from constants import (ACCURACIES_PATH, COND_ACC_PATH, RTs_PATH, COND_RTs_PATH, ACC_UNSHIFTED_PATH, TRIAL_AVERAGE_PATH)
from utils.utils_preprocessing import (get_mean_run_measure, get_mean_block_measure, get_run_error, get_mean_subj_measure, get_subj_error, get_block_error)
from utils.utils_plotting import (graph_all_participants, graph_mean_run, graph_performance_individual_participants, barplot_mean_subject, barplot_mean_block, plot_cond_measure, graph_mean_run_layered, plot_by_trial, roving_window)
```

# Imports
Importing all the preprocessed data with accuracies and reaction times

```python
accuracies = pd.read_csv(ACCURACIES_PATH, header=None)
cond_acc = pd.read_csv(COND_ACC_PATH, header=None)
acc_unshifted = pd.read_csv(ACC_UNSHIFTED_PATH, header=None)
RT = pd.read_csv(RTs_PATH, header=None)
cond_RT = pd.read_csv(COND_RTs_PATH, header=None)
trials_average = pd.read_csv(TRIAL_AVERAGE_PATH, header=None)
```

# Graphs
#### All participants accuracy over all runs

```python
graph_all_participants(measure=accuracies)
```

#### Performance of individual subjects

```python
graph_performance_individual_participants(measure=accuracies)
```

#### Mean accuracy over all runs

```python
mean_run_accuracies = get_mean_run_measure(measure=accuracies)
run_error_acc = get_run_error(measure=accuracies)
graph_mean_run(mean_run_measure=mean_run_accuracies, run_error=run_error_acc, title="Mean Accuracy over all 12 runs", yaxis="Accuracy")
```

#### Accuracy of a single participant over all runs

```python
graph_mean_run(mean_run_measure=accuracies.iloc[1], run_error=run_error_acc, title="Mean Accuracy over all 12 runs", yaxis="Accuracy")
```

#### Mean accuracy comparing day 1 with day 2

```python
mean_run_acc_unshifted = get_mean_run_measure(measure=acc_unshifted)
run_error_acc_uns = get_run_error(measure=acc_unshifted)
graph_mean_run(mean_run_measure=mean_run_acc_unshifted, run_error=run_error_acc_uns,
               title="Mean Accuracy comparing the days", yaxis="Accuracy")
```

#### Mean RT over all runs

```python
mean_run_RT = get_mean_run_measure(measure=RT)
run_error_RT = get_run_error(measure=RT)
graph_mean_run(mean_run_measure=mean_run_RT, run_error=run_error_RT, title="Mean RT over all 12 runs", yaxis="Reaction Time")
```
